```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotAndExport(i, Z):
    fig = plt.figure(figsize=(15,9))
    x_vals = np.arange(2, 8.1, 0.1)
    y_vals = np.arange(30, 121, 1)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = Z
    contour = plt.contourf(X, Y, Z, 10, cmap='coolwarm')
    plt.colorbar(contour)
    plt.title('Sample ' + str(i), fontsize=20)
    fig.savefig('images/sample(' + str(i) + ').png')
```

```python
for i in range(100):
    filename = './data/Project 2 sample ('+ str(i+1) +').csv'
    df = pd.read_csv(filename, header=None)
    dataArray = np.asarray(df)
    plotAndExport(i+1, dataArray)
    print('save sample ' + str(i+1))
```
